```python
evaluator = Evaluator()

def evaluateResults(stances, prediction, decimals=4):
    score = evaluator.getScore(stances, prediction)
    score_percentage = evaluator.getScorePercentage(
        stances, prediction)
    is_related_correctness = evaluator.getIsRelatedCorrectness(
        stances, prediction)
    unrelated_correctness = evaluator.getUnrelatedCorrectness(
        stances, prediction)
    related_correctness = evaluator.getRelatedCorrectness(
        stances, prediction)
    relationship_correctness = evaluator.getRelationshipCorrectness(
        stances, prediction)
    print('Results:')
    print('    score:', score)
    print('    score_percentage: {:.{precision}f}'. \
        format(score_percentage, precision=decimals))
    print('    is_related_correctness: {:.{precision}f}' \
        .format(is_related_correctness, precision=decimals))
    print('    unrelated_correctness: {:.{precision}f}' \
        .format(unrelated_correctness, precision=decimals))
    print('    related_correctness: {:.{precision}f}' \
        .format(related_correctness, precision=decimals))
    print('    relationship_correctness: {:.{precision}f}' \
        .format(relationship_correctness, precision=decimals))
    return
```

```python
import numpy as np
from matplotlib_venn import venn3

def venn_diagram(competition_all_teams_data):
    labels=['A', 'B', 'C']
    stances = competition_all_teams_data['Stance']
    athene_stances = competition_all_teams_data['AtheneStance']
    talos_stances = competition_all_teams_data['TalosStance']
    ucl_stances = competition_all_teams_data['UclStance']
    athene_correct = athene_stances == stances
    talos_correct = talos_stances == stances
    ucl_correct = ucl_stances == stances
    athene_wrong = np.logical_not(athene_correct)
    talos_wrong = np.logical_not(talos_correct)
    ucl_wrong = np.logical_not(ucl_correct)

    only_athene = np.sum(athene_wrong & talos_correct & ucl_correct)
    only_talos = np.sum(athene_correct & talos_wrong & ucl_correct)
    only_ucl = np.sum(athene_correct & talos_correct & ucl_wrong)
    only_athene_talos = np.sum(athene_wrong & talos_wrong & ucl_correct)
    only_athene_ucl = np.sum(athene_wrong & talos_correct & ucl_wrong)
    only_talos_ucl = np.sum(athene_correct & talos_wrong & ucl_wrong)
    athene_talos_ucl = np.sum(athene_wrong & talos_wrong & ucl_wrong)

    venn3(subsets=(only_athene, only_talos, only_ucl, only_athene_talos,\
         only_athene_ucl, only_talos_ucl, athene_talos_ucl), set_labels=labels,
         normalize_to=10)


venn_diagram(competition_all_teams_data)
```
